üñ• Nettoyage des donn√©es

```python
# Remplacer les valeurs manquantes par des indicateurs sp√©cifiques
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)

# Convertir date_added en type date
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# R√©examen des donn√©es apr√®s nettoyage
df.info()
```

üñ• Distribution des types de contenu

```python
import matplotlib.pyplot as plt

# Distribution des types de contenu
content_type_counts = df['type'].value_counts()

# Visualisation
plt.figure(figsize=(8, 6))
content_type_counts.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Distribution des Types de Contenu')
plt.xlabel('Type de Contenu')
plt.ylabel('Nombre de Titres')
plt.xticks(rotation=0)
plt.show()
```

üñ• Analyse des genres et cat√©gories

```python
# Compter les genres
genre_counts = df['listed_in'].str.split(', ').explode().value_counts().head(10)

# Visualisation
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Genres de Contenu')
plt.xlabel('Genre')
plt.ylabel('Nombre de Titres')
plt.xticks(rotation=80)
plt.show()
```

üñ• Analyse par pays d'origine

```python
# Compter les pays
country_counts = df['country'].str.split(', ').explode().value_counts().head(10)

# Visualisation
plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='lightseagreen')
plt.title('Top 10 Pays d\'Origine des Titres')
plt.xlabel('Pays')
plt.ylabel('Nombre de Titres')
plt.xticks(rotation=80)
plt.show()
```

üñ• Analyse des dur√©es et des saisons

```python
# S√©parer les films et les s√©ries
movies = df[df['type'] == 'Movie']
tv_shows = df[df['type'] == 'TV Show']
```

```python
print(movies.shape)
movies.head()
```

```python
# Filtrer les valeurs "Unknown" avant de convertir la colonne duration en entier
movies = movies[movies['duration'] != 'Unknown']

# Distribution des dur√©es des films
movies['duration'] = movies['duration'].str.replace(' min', '').astype(int)

plt.figure(figsize=(10, 6))
movies['duration'].plot(kind='hist', bins=30, color='lightskyblue')
plt.title('Distribution des Dur√©es des Films')
plt.xlabel('Dur√©e (minutes)')
plt.ylabel('Nombre de Films')
plt.show()
```

```python
tv_shows = tv_shows[tv_shows['duration'] != 'Unknown']

# Distribution des saisons des s√©ries
tv_shows['duration'] = tv_shows['duration'].str.replace(' Season', '').str.replace('s', '').astype(int)

plt.figure(figsize=(10, 6))
tv_shows['duration'].plot(kind='hist', bins=10, color='lightcoral')
plt.title('Distribution du Nombre de Saisons des S√©ries')
plt.xlabel('Nombre de Saisons')
plt.ylabel('Nombre de S√©ries')
plt.show()
```

üñ• Distribution des dates d'ajout et des ann√©es de sortie

```python
# Distribution des dates d'ajout
plt.figure(figsize=(10, 6))
df['date_added'].dt.year.value_counts().sort_index().plot(kind='bar', color='mediumseagreen')
plt.title('Distribution des Dates d\'Ajout')
plt.xlabel('Ann√©e')
plt.ylabel('Nombre de Titres Ajout√©s')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution des ann√©es de sortie
plt.figure(figsize=(18, 6))
df['release_year'].value_counts().sort_index().plot(kind='bar', color='mediumpurple')
plt.title('Distribution des Ann√©es de Sortie')
plt.xlabel('Ann√©e')
plt.ylabel('Nombre de Titres Sortis')
plt.xticks(rotation=45)
plt.show()
```
